```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as npr
import math

def matrizz(a,v):

    r=np.matmul(a.data,v)
    print(r)
    return r


a= np.matrix([[1,3], [-2,5]])
x= np.arange(-1,6,1)
v= np.array([3,4])
r= matrizz(a,v)


plt.plot(x,r[0])
plt.plot(x,r[1])

plt.ylabel("parallel")
plt.show()
    
```
